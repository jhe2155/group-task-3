```python
import pandas as pd

co2 = pd.read_csv('co2_emissions.csv')
cars = pd.read_csv('state_cars.csv')

co2['id'] = cars['id'].astype(int)
cars['id'] = cars['id'].astype(int)

row1 = cars.shape[0]

for i in range(row1):
    cars.at[i,'Cars'] = cars['Cars'][i].replace(',','')
    
cars['Cars'] = cars['Cars'].astype(int)

row2 = co2.shape[0]

co2['2018'] = co2['2018'].astype(int)
```

```python
import altair as alt
from vega_datasets import data

states = alt.topo_feature(data.us_10m.url, 'states')

alt.Chart(states, title='Number of Registered Cars By State').mark_geoshape().encode(
    alt.Color('Cars:Q',title='Number of Cars')
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(cars, 'id', ['Cars'])
).project(
    type='albersUsa'
).properties(
    width=500,
    height=300
)
```

```python
alt.Chart(states, title='CO2 Emissions by State').mark_geoshape().encode(
    alt.Color('2018:Q',title='Million Metric Tons Produced', scale=alt.Scale(scheme='inferno')),
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(co2, 'id', ['2018'])
).project(
    type='albersUsa'
).properties(
    width=500,
    height=300
)
```
